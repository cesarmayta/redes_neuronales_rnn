## 1.1 EJEMPLO SENCILLO CON DATOS SIMULADOS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import scipy.stats
```

# EJEMPLO DE DATOS PARA UNA RELACIÓN LINEAL

```python
x = np.linspace(0, 20, 100)  # 100 puntos entre 0 y 20
y = 4 * x + 7                # Relación lineal: y = 4x + 7

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = 4x + 7', color='orange')
plt.title('Ejemplo de relación lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.show()
```

La fórmula de una relación lineal es:

$$
y = mx + b
$$

Donde:
- \( m \) es la pendiente
- \( b \) es la ordenada al origen (intersección con el eje y)

Ejemplo:

$$
y = 4x + 7
$$


### REALIZAMOS LA REGRESIÓN LINEAL SIMPLE

```python
result = scipy.stats.linregress(x,y)
```

### MOSTRAMOS RESULTADOS DE LA REGRESIÓN LINEAL

```python
print(f'Pendiente(slope): {result.slope}')
print(f'Intercepto(intercept): {result.intercept}')
print(f'Coeficiente de Correlación(rvalue) {result.rvalue}')
print(f'Valor p(pvalue) : {result.pvalue}')
print(f'Error stándar de la pendiente (stderr) : {result.stderr}')
```

```python
y_pred = result.slope * x + result.intercept
```

### GRAFICAMOS LA REGRESIÓN LINEAL

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Datos Observados')
plt.plot(x,y_pred,color='red',label='Línea de regresión')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión lineal con scipy')
plt.legend()
plt.show()
```

### PROBAMOS NUESTRA PREDICCIÓN

```python
x_new = 21
y_new_pred = result.slope * x_new + result.intercept
print(f'predicción para x = {x_new}: y = {y_new_pred}')
```

### GRAFICAMOS EL VALOR DE PREDICCIÓN

```python

m = result.slope
b = result.intercept
y_pred_x = m * x + b
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='datos observador')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y_pred_x,color='red',label='Linea de regresión')
plt.scatter(x_new,y_new_pred,color='green',marker='o',s=100,label=f'predicción para x={x_new}')

plt.legend()
plt.show()
```

# DATOS PARA UNA RELACIÓN CUADRATICA

```python
x = np.linspace(0, 100, 100)  # Valores de x entre -10 y 10
y = 2 * x**2 + 3 * x + 5       # Relación cuadrática

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = 4x + 7', color='orange')
plt.title('Ejemplo de relación lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.show()
```

La fórmula de una relación cuadrática es:

$$
y = ax^2 + bx + c
$$


```python
result = scipy.stats.linregress(x,y)
y_pred = result.slope * x + result.intercept
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Datos Observados')
plt.plot(x,y_pred,color='red',label='Línea de regresión')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión lineal con scipy')
plt.legend()
plt.show()
```
